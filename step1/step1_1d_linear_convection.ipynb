```python
# Standard header
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
# Physical parameter
c = 1   # Wave speed

# Numerical parameters
nx = 41
L = 2
Tf = 0.625
dt = 0.025

# Driven numerical parameters
dx = L/(nx-1)
nt = int(Tf/dt)
x = numpy.linspace(0,L,nx)
```

```python
# Plotting function
def plot(x,y):
    pyplot.figure(figsize=(16,8))
    pyplot.grid()
    pyplot.plot(x,y);
```

```python
# Initial condition set and display
u0 = numpy.ones(nx)
u0[int(.5/dx):int(1/dx)+1] = 2
plot(x,u0)
```

```python
# Explicit forward time-step
u = numpy.copy(u0)
for it in range(nt):
    un = numpy.copy(u)
    u[1:] = un[1:] - c*dt/dx*(u[1:] - u[:-1])
    
# Display the transported solution
plot(x,u)
```
